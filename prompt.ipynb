**Context**

**Data**

This data comes to us from the UCI Machine Learning repository and was collected via a survey on Amazon Mechanical Turk. There are five different types of coupons -- less expensive restaurants (under \$20), coffee houses, carry out & take away, bar, and more expensive restaurants (\$20 - $50).

2. Investigate the dataset for missing or problematic data.

```python
data.isna().sum()
```

3. Decide what to do about your missing data -- drop, replace, other...

```python
# most of the values for car is missing, so we can drop that column
data_without_car = data.drop(columns=['car'])

# for the rest of the columns, we can fill the missing values with the mode of the column
columns_with_na = data_without_car.columns[data_without_car.isna().sum() > 0]
for column in columns_with_na:
    mode_value = data_without_car[column].mode()[0]
    data_without_car[column] = data_without_car[column].fillna(mode_value)

cleaned_data = data_without_car
```

4. What proportion of the total observations chose to accept the coupon?



```python
total_rows = cleaned_data.shape[0]
print(f'Total number of rows in the cleaned dataset: {total_rows}')
total_accepted_coupons = cleaned_data[cleaned_data['Y'] == 1].shape[0]
print(f'Total number of accepted coupons: {total_accepted_coupons}')
acceptance_rate = (total_accepted_coupons / total_rows) * 100
print(f'Acceptance rate: {acceptance_rate:.2f}%')
```

**Investigating the Bar Coupons**

Now, we will lead you through an exploration of just the bar related coupons.  

1. Create a new `DataFrame` that contains just the bar coupons.


```python

bar_coupons = cleaned_data[cleaned_data['coupon'] == 'Bar']
bar_coupons.info()
```

2. What proportion of bar coupons were accepted?


```python
total_rows = bar_coupons.shape[0]
print(f'Total number of rows in the bar dataset: {total_rows}')
total_accepted_coupons = bar_coupons[bar_coupons['Y'] == 1].shape[0]
print(f'Total number of accepted coupons: {total_accepted_coupons}')
acceptance_rate = (total_accepted_coupons / total_rows) * 100
print(f'Acceptance rate: {acceptance_rate:.2f}%')

#Trying an alternative way to calculate acceptance rate
print(f'Acceptance rate for Bar coupons: {bar_coupons["Y"].mean() * 100:.2f}%')
```

3. Compare the acceptance rate between those who went to a bar 3 or fewer times a month to those who went more.


```python
# getting the unique values of the 'bar' column
print(bar_coupons['Bar'].unique())

three_or_fewer = bar_coupons['Bar'].isin(['never', 'less1', '1~3'])
more_than_three = bar_coupons['Bar'].isin(['4~8', 'gt8'])

rate_three_or_fewer = bar_coupons.loc[three_or_fewer, 'Y'].mean()
rate_more_than_three = bar_coupons.loc[more_than_three, 'Y'].mean()

print(f'Acceptance (bar visits <= 3/month): {rate_three_or_fewer:.2%}')
print(f'Acceptance (bar visits > 3/month):  {rate_more_than_three:.2%}')
```

4. Compare the acceptance rate between drivers who go to a bar more than once a month and are over the age of 25 to the all others.  Is there a difference?


```python
print(bar_coupons['age'].unique())

is_bar_more_than_once = bar_coupons['Bar'].isin(['1~3', '4~8', 'gt8'])
is_over_25 = bar_coupons['age'].isin(['26', '46', '31', '41', '50plus','36'])

target_group = is_bar_more_than_once & is_over_25
target_rate = bar_coupons.loc[target_group, 'Y'].mean()
other_rate = bar_coupons.loc[~target_group, 'Y'].mean()

print(f'Acceptance (bar >1/month and age >25): {target_rate:.2%}')
print(f'Acceptance (all others):               {other_rate:.2%}')
```

5. Use the same process to compare the acceptance rate between drivers who go to bars more than once a month and had passengers that were not a kid and had occupations other than farming, fishing, or forestry.


```python
bar_more_than_once = bar_coupons['Bar'].isin({'1~3', '4~8', 'gt8'})
no_kid_passenger = bar_coupons['passanger'].ne('Kid(s)')
not_farming = bar_coupons['occupation'].ne('Farming Fishing & Forestry')

group_a = bar_more_than_once & no_kid_passenger & not_farming

rate_group_a = bar_coupons.loc[group_a, 'Y'].mean()
rate_others_a = bar_coupons.loc[~group_a, 'Y'].mean()

print(f'Acceptance (specified group): {rate_group_a:.2%}')
print(f'Acceptance (all others):      {rate_others_a:.2%}')
```

6. Compare the acceptance rates between those drivers who:

- go to bars more than once a month, had passengers that were not a kid, and were not widowed *OR*
- go to bars more than once a month and are under the age of 30 *OR*
- go to cheap restaurants more than 4 times a month and income is less than 50K.



```python
condition_1 = (
    bar_coupons['Bar'].isin(['1~3', '4~8', 'gt8'])
    & (bar_coupons['passanger'] != 'Kid(s)')
    & (bar_coupons['maritalStatus'] != 'Widowed')
)

condition_2 = (
    bar_coupons['Bar'].isin(['1~3', '4~8', 'gt8'])
    & (bar_coupons['age'].isin(['21','26','below21']))
)

low_income = ['Less than $12500', '$12500 - $24999', '$25000 - $37499', '$37500 - $49999']
condition_3 = (
    bar_coupons['RestaurantLessThan20'].isin(['4~8', 'gt8'])
    & bar_coupons['income'].isin(low_income)
)

combined_group = condition_1 | condition_2 | condition_3
combined_rate = bar_coupons.loc[combined_group, 'Y'].mean()
others_rate = bar_coupons.loc[~combined_group, 'Y'].mean()

print(f'Acceptance (combined conditions): {combined_rate:.2%}')
print(f'Acceptance (all others):          {others_rate:.2%}')
```

7.  Based on these observations, what do you hypothesize about drivers who accepted the bar coupons?

```python
print(
    'Hypothesis: Drivers who are socially active (visit bars more often), not traveling with kids, and younger or mid-age adults are more likely '
    'to accept bar coupons. Lifestyle context (companions, habits, and income) appears to strongly influence acceptance behavior.'
)
```
